# Intersección de superficies

Consideramos el conjunto de puntos que satisfacen simultáneamente:

$$
x^2 + y^2 = 1
$$

$$
z = y
$$

El objetivo es describir la curva de intersección entre estas dos superficies.

## Parametrización del conjunto de intersección

La ecuación
$$
x^2 + y^2 = 1
$$
describe una circunferencia unitaria en el plano $xy$.

Por lo tanto, sus puntos pueden escribirse como:
$$
x = \cos t, \quad y = \sin t
$$

donde $t$ es un parámetro angular.

---

El plano impone la condición:
$$
z = y
$$

Sustituyendo $y = \sin t$, obtenemos:
$$
z = \sin t
$$

---

Por lo tanto, el conjunto de puntos de la intersección está dado por:
$$
(x,y,z) = (\cos t, \sin t, \sin t)
$$

Esta expresión describe todos los puntos que satisfacen simultáneamente:
$$
x^2 + y^2 = 1 \quad \text{y} \quad z = y
$$

```python
import numpy as np
import plotly.graph_objects as go

# curva
t = np.linspace(0, 4*np.pi, 400)
x = np.cos(t)
y = np.sin(t)
z = np.sin(t)

# cilindro
theta = np.linspace(0, 2*np.pi, 50)
z_vals = np.linspace(-1.5, 1.5, 50)
Theta, Z = np.meshgrid(theta, z_vals)

Xc = np.cos(Theta)
Yc = np.sin(Theta)

# plano z = y
x_vals = np.linspace(-1.5, 1.5, 50)
y_vals = np.linspace(-1.5, 1.5, 50)
Xp, Yp = np.meshgrid(x_vals, y_vals)
Zp = Yp

# figura
fig = go.Figure()

# curva
fig.add_trace(go.Scatter3d(
    x=x, y=y, z=z,
    mode='lines',
    name='Curva'
))

# cilindro
fig.add_trace(go.Surface(
    x=Xc, y=Yc, z=Z,
    opacity=0.3,
    showscale=False,
    name='Cilindro'
))

# plano
fig.add_trace(go.Surface(
    x=Xp, y=Yp, z=Zp,
    opacity=0.3,
    showscale=False,
    name='Plano z=y'
))

# layout
fig.update_layout(
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='z'
    ),
    width=700,
    height=700
)

fig.show()
```

## Interpretación del parámetro

El parámetro $t$ determina la posición del punto sobre la curva.

Para cada valor de $t$, obtenemos un punto:
$$
(x,y,z) = (\cos t, \sin t, \sin t)
$$

Al variar $t$, se recorre completamente la curva de intersección.

---

Observación:

- $x$ y $y$ describen un movimiento circular en el plano
- $z$ reproduce exactamente el valor de $y$

Esto implica que la curva se obtiene elevando la circunferencia según la regla $z = y$.

---

## Vector posición

Cada punto de la curva puede interpretarse como el extremo de un vector que parte del origen. En efecto, si

$$
P(t) = (\cos t, \sin t, \sin t),
$$

entonces el vector posición asociado es

$$
\vec{r}(t) = (\cos t, \sin t, \sin t).
$$

Este vector representa la ubicación del punto en el espacio en función del parámetro.

---

Interpretación geométrica:

- El vector $\vec{r}(t)$ une el origen con un punto de la curva.
- A medida que $t$ varía, el extremo del vector describe exactamente la trayectoria observada.
- Por tanto, la curva puede entenderse como el conjunto de extremos de los vectores $\vec{r}(t)$.

---

Relación con la animación:

En la visualización dinámica:

- el punto móvil corresponde al extremo del vector,
- la flecha dibujada corresponde al vector posición,
- el movimiento del vector ilustra cómo cambia la posición en función de $t$.

---

Conclusión:

La parametrización no solo describe una colección de puntos, sino un proceso dinámico en el cual cada valor de $t$ determina un vector de posición distinto en $\mathbb{R}^3$, cuyo extremo recorre la curva de intersección.

## Verificación mediante puntos del círculo unitario

Evaluamos la parametrización en ángulos notables:

| $t$ | $x = \cos t$ | $y = \sin t$ | $z = y$ | Punto $(x,y,z)$ |
|----|-------------|-------------|--------|----------------|
| $0$ | $1$ | $0$ | $0$ | $(1,0,0)$ |
| $\frac{\pi}{6}$ | $\frac{\sqrt{3}}{2}$ | $\frac{1}{2}$ | $\frac{1}{2}$ | $\left(\frac{\sqrt{3}}{2},\frac{1}{2},\frac{1}{2}\right)$ |
| $\frac{\pi}{4}$ | $\frac{\sqrt{2}}{2}$ | $\frac{\sqrt{2}}{2}$ | $\frac{\sqrt{2}}{2}$ | $\left(\frac{\sqrt{2}}{2},\frac{\sqrt{2}}{2},\frac{\sqrt{2}}{2}\right)$ |
| $\frac{\pi}{2}$ | $0$ | $1$ | $1$ | $(0,1,1)$ |
| $\pi$ | $-1$ | $0$ | $0$ | $(-1,0,0)$ |
| $\frac{3\pi}{2}$ | $0$ | $-1$ | $-1$ | $(0,-1,-1)$ |

```python
import numpy as np
import plotly.graph_objects as go

# curva
t = np.linspace(0, 4*np.pi, 200)
x = np.cos(t)
y = np.sin(t)
z = np.sin(t)

# puntos notables
t_pts = np.array([0, np.pi/6, np.pi/4, np.pi/2, np.pi, 3*np.pi/2])
x_pts = np.cos(t_pts)
y_pts = np.sin(t_pts)
z_pts = np.sin(t_pts)

labels = [
    "(1,0,0)",
    "(√3/2,1/2,1/2)",
    "(√2/2,√2/2,√2/2)",
    "(0,1,1)",
    "(-1,0,0)",
    "(0,-1,-1)"
]

# desplazamiento hacia la derecha
offset = 0.15
x_text = x_pts + offset

# frames
frames = []
t_frames = np.linspace(0, 4*np.pi, 100)

for ti in t_frames:
    xp = np.cos(ti)
    yp = np.sin(ti)
    zp = np.sin(ti)

    frames.append(go.Frame(
        data=[
            go.Scatter3d(x=[xp], y=[yp], z=[zp]),
            go.Scatter3d(x=[0, xp], y=[0, yp], z=[0, zp])
        ],
        traces=[2, 3],
        name=str(ti)
    ))

# figura base
fig = go.Figure(
    data=[
        # curva
        go.Scatter3d(x=x, y=y, z=z, mode='lines', name='Curva'),

        # puntos notables
        go.Scatter3d(
            x=x_pts, y=y_pts, z=z_pts,
            mode='markers',
            name='Puntos notables'
        ),

        # etiquetas desplazadas
        go.Scatter3d(
            x=x_text, y=y_pts, z=z_pts,
            mode='text',
            text=labels,
            name='Etiquetas'
        ),

        # punto móvil
        go.Scatter3d(
            x=[x[0]], y=[y[0]], z=[z[0]],
            mode='markers',
            marker=dict(size=6),
            name='Punto'
        ),

        # vector
        go.Scatter3d(
            x=[0, x[0]], y=[0, y[0]], z=[0, z[0]],
            mode='lines',
            name='Vector'
        )
    ],
    frames=frames
)

# slider
sliders = [dict(
    steps=[
        dict(method="animate",
             args=[[f.name], {"mode": "immediate"}],
             label=f"{float(f.name):.2f}")
        for f in frames
    ]
)]

# botones
updatemenus = [dict(
    type="buttons",
    buttons=[
        dict(label="Play", method="animate", args=[None]),
        dict(label="Pause", method="animate", args=[[None], {"mode": "immediate"}])
    ]
)]

# layout
fig.update_layout(
    scene=dict(
        xaxis=dict(range=[-1.5, 1.5], title='x'),
        yaxis=dict(range=[-1.5, 1.5], title='y'),
        zaxis=dict(range=[-1.5, 1.5], title='z'),
        aspectmode='cube'
    ),
    width=700,
    height=700,
    sliders=sliders,
    updatemenus=updatemenus
)

fig.show()
```
